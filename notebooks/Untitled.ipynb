```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import missingno as msno
```

    ##chemin vers les données brutes :

```python
data_path = "../data/raw"
```

    ###load all datafiles via a loop & showing basic info :

```python
for file in os.listdir(data_path):
    if file.endswith(".csv"):
        print("=" * 80)
        print(f" file analysed: {file}")
        print("=" * 80)
        file_path = os.path.join(data_path, file)
        df = pd.read_csv(file_path, header=1)
        #mettre en forme les datasets pour qu'ils aient les bons noms de colonnes
        #df2 = df.iloc[1:].reset_index(drop=True)


        print("\n Aperçu des premières lignes :")
        print(df.head())
        
        print("\n Info sur les colonnes :")
        print(df.info())
        
        print("\n Types de colonnes :")
        print(df.dtypes)
        
        print("\n Valeurs manquantes :")
        print(df.isnull().sum())
        
        print("\n Statistiques descriptives (variables numériques) :")
        print(df.describe())
        
        print("\n Statistiques descriptives (variables qualitatives) :")
        print(df.describe(include=['object']))
        
        
        print("\n\n\n")  # sauts de ligne 

        #étude des valeurs manquantes
        print(f"\n===  Etude des valeurs manquantes : {file} ===")
        
        # 1. Barplot des valeurs manquantes
        missing = df.isnull().sum()
        if missing.sum() == 0:
            print(" Aucune valeur manquante.")
        else:
            missing[missing > 0].sort_values(ascending=False).plot(
                kind='bar',
                figsize=(10, 4),
                title=f"Valeurs manquantes par colonne ({file})",
                color='tomato'
            )
            plt.ylabel("Nombre de NaN")
            plt.xticks(rotation=45, ha="right")
            plt.tight_layout()
            plt.grid(True)
            plt.show()
        
        # 2. Schéma avec la matrice des valeurs manquantes
        if missing.sum() > 0:
            print(" missingno.matrix()=")
            msno.matrix(df)
            plt.show()
```
